In this part, we will try to classify music using their audio files. The audio was retrieved using the Spotify API and the Spotipy Python library. See our notebooks for that: https://github.com/Ferdinand-Genans/SpotifyProject
In order to classify music using a CNN, we had to convert the music into audio.
We used the "chromogram" representations of the music.

A chromagram is a visual representation of the chromatic content of a piece of music. It is similar to a spectrogram, which shows the frequency content of a signal, but instead of showing frequencies on the vertical axis, a chromagram shows the 12 pitch classes of the chromatic scale.
The chromagram is  constructed by taking the short-time Fourier transform (STFT) of a piece of music, and then calculating the energy of each pitch class at each time frame.

To perform this representation, we used the librosa library.

Let's give an example: 

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
from scipy.io.wavfile import read
import librosa
import librosa.display
import os
from PIL import Image
import pathlib
import csv
import warnings
warnings.filterwarnings('ignore')
```

```python
path = "Audio_processing/genres_original/blues/blues.00000.wav"
input_data = read(path)
audio = input_data[1]
plt.plot(audio)
plt.ylabel("Amplitude")
plt.xlabel("Time")
plt.title("Music mp3 example")
plt.show()
```

```python
y, sr = librosa.load(path)
C = librosa.feature.chroma_cqt(y=y, sr=sr)
fig = plt.Figure(figsize=(6,4))
img = librosa.display.specshow(C, y_axis='chroma', x_axis='time')
plt.colorbar(img)
plt.title('Chromogram')
plt.show()
```
