```python
# dataset link: https://www.kaggle.com/datasets/maharshipandya/-spotify-tracks-dataset

import pandas as pd 
df = pd.read_csv("dataset.csv", index_col=0) #loading the dataset 
df = df.dropna() #only one music
genres = ['classical','country', 'house', 'metal', 'k-pop', 'jazz', 'rock', 'pop', 'r-n-b', 'reggae']
df = df.loc[df.track_genre.isin(genres)]
df.reset_index(drop = True, inplace = True) #reset index
df.to_csv("dataset_clean.csv", index = False)
```
