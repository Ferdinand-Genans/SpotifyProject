La librairie librosa se charge de ces transformations, en utilisant le signal audio de notre mp3, puis en utilisant notamment des transormers de Fourier FFT.

Illustrons cela par un exemple de musique. 

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
from scipy.io.wavfile import read
import librosa
import librosa.display
import os
from PIL import Image
import pathlib
import csv
import warnings
warnings.filterwarnings('ignore')
```

```python
path = "Audio_processing/genres_original/blues/blues.00000.wav"
input_data = read(path)
audio = input_data[1]
plt.plot(audio)
plt.ylabel("Amplitude")
plt.xlabel("Temps")
plt.title("Exemple de son")
plt.show()
```

```python
y, sr = librosa.load(path)
plt.figure(figsize=(6,4))
D = librosa.amplitude_to_db(librosa.stft(y), ref=np.max)
librosa.display.specshow(D, y_axis='log')
plt.colorbar(format='%+2.0f dB')
plt.title('Log-frequency power spectrogram')
plt.show()
C = librosa.feature.chroma_cqt(y=y, sr=sr)
fig = plt.Figure()
```

```python
C = librosa.feature.chroma_cqt(y=y, sr=sr)
fig = plt.Figure(figsize=(6,4))
img = librosa.display.specshow(C, y_axis='chroma', x_axis='time')
plt.colorbar(img)
plt.title('Chromogram')
plt.show()
```
